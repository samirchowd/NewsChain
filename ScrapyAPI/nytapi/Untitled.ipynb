```python
import random


def generateBoard(words):
    result = []
    
    for word in words:
        result.append([word, False])
        result.append([word, False])
        
    random.shuffle(result)
    
    return result

def displayBoard(board):
    
    for card in range(len(board)):
        if board[card][1] == False:
            print(str(card) + ": ???")
        else:
            print(str(card) + ": " + board[card][0])
        
    return

def getFlippedCards(board):
    result = []
    
    for card in range(len(board)):
        
        if board[card][1] == True:
            result.append(card)
            
    return result

def pickIndex(board):
    result = -1
    while True:
        try:
            index = int(input("Pick a card index: "))
            if index >= 0 and index < len(board):
                if board[index][1] == False:
                    board[index][1] = True
                    result = index
                    break
                else:
                    print("You've already flipped that card. Pick a different card.")
            else:
                print("Out of range. Pick a valid card.")
        except:
            print("That's not an integer!")
    return result

def playMemoryGame(words):
    print("Welcome to Alicia's Memory Game!! :D")
    
    board = generateBoard(words)
    displayBoard(board)
    
    turnCount = 0
    
    while getFlippedCards(board) != list(range(len(board))):
        pick1 = pickIndex(board)
        displayBoard(board)
        pick2 = pickIndex(board)
        displayBoard(board)
        
        if board[pick1][0] == board[pick2][0]:
            print("Good guess!")
        else:
            board[pick1][1] = False
            board[pick2][1] = False
            print("Nice try!")
            
        print("---")
        
        turnCount += 1
        
    print("Congratulations! You won in " + str(turnCount) + " moves!")
    return

words = ["shampoo", "conditioner", "body wash"]
playMemoryGame(words)
```
